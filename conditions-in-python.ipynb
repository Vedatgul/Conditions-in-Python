****Control structures in programming languages ensure that given codes work according to the situation by checking whether a given condition is fulfilled. In this way, the flow of the program is directed. We perform this operation with “if, elif and else” statements.****

**The following operators are used in conditional structures.**

**if:** *If the condition in the if condition is met, the relevant operations are performed.*

```python
### If a is equal to 1, the condition is satisfied.
a=1
if (a==1):
  print(a)
```

**elif:** *If the if condition is not met, elif conditions are checked.*

```python
### If a is equal to 2, the elif block is active.
a=2
if (a==1):
  print(a)

elif (a==2):
  print(a)
```

**else:** *If any condition is not met, operations under the else structure are performed.*

```python
### If a is equal to 3, the else block is active.
a=3
if (a==1):
  print(a)

elif (a==2):
  print(a)

else:
  print(a)
```
